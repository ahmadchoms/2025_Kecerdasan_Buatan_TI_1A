```python
import pandas as pd

# Membuat Dataframe
df = pd.DataFrame(data = {
    "Name": ["Ali", "Joko", "Adi"],
    "Umur": [12, 13, 15],
    "Kelas": [6, 7, 8]
    })
df.columns
df.columns.values[0] = "Nama Singkat"
print(df)
```

```python
data = {
    "calories": [420, 380, 390],
    "duration": [50, 40, 45]
}

df = pd.DataFrame(data)
df

# Mengakses Elemen pada DataFrame
print("df.iloc[0:2, [0, 1]]")
print(df.iloc[0:2, [0, 1]])
print("\n")

print("df.iloc[0:2, 0:2]")
print(df.iloc[0:2, 0:2])
print("\n")

print("df.iloc[:, 0:2]")
print(df.iloc[:, 0:2])
print("\n")

print("df.loc[0:2, : 'calories']")
print(df.loc[0:2, : 'calories'])
print("\n")

print("df.loc[0:2, ['calories', 'duration']]")
print(df.loc[0:2, ['calories', 'duration']])
print("\n")

print("df['calories']")
print(df['calories'])
print("\n")

print("df['calories'][0]")
print(df['calories'][0])
print("\n")

print("df[['calories']]")
print(df[['calories']])
print("\n")

print("df[['calories']].loc[[0, 1]]")
print(df[['calories']].loc[[0, 1]])
print("\n")
```

```python
# Memberi Nama pada Index

df = pd.DataFrame(data, index = ['day1', 'day2', 'day3'])
print(df)
```
